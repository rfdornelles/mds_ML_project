```python
# Importing libraries
import pandas as pd
import numpy as np
import csv
import matplotlib as plt
from datetime import datetime as dt
import datetime as dt
```

```python
# Reading file
data = pd.read_csv("data-raw/Deaths_Germany_years_nationality_sex_age_years.csv", encoding= 'unicode_escape')
```

```python
# Printing dataframe
data.head()
```

```python
# Renaming columns
data.rename(columns = {"Unnamed: 0" : "Year", "Unnamed: 1" : "Age",
                       "Germans": "Male_Germans", "Foreigners" : "Male Foreigners",
                       "Total" : "Total Male",
                      "Germans.1" : "Female_Germans", "Foreigners.1" : "Female Foreigners",
                       "Total.1" : "Total Fem",
                      "Germans.2" : "Total_Germans", "Foreigners.2" : "Total Foreigners",
                      "Total.2" : "Total"}, inplace = True)

# Printing
data.head()
```

```python
# Setting index to 1 
data.index += 1

# Printing
data.head(50)
```

```python
# Selecting columns to analyze
data = data.loc[:, ["Year", "Age", "Male_Germans", "Female_Germans"]]

# Printing
data.head()
```

```python
# Cleaning 'Age' column
data['Age'] = data['Age'].str.replace('years', '')

# Printing
data.head()
```

```python
# Cleaning 'Age' column
data['Age'] = data['Age'].str.replace(' s', '')

# Printing
data.head()
```

```python
# Cleaning 'Age' column
data['Age'] = data['Age'].str.replace(' and over', '')

# Printing
data.head()
```

```python
# Deleting all rows with 'Total'
data = data[data.Age != 'Total']

# Printing
data.head()
```

```python
# Deleting all rows with 'age unknown'
data = data[data.Age != 'age unknown']

# Printing
data.head()
```

```python
# Cleaning 'Age' column
data['Age'] = data['Age'].str.replace(' year', '')

# Printing
data.head(100)
```

```python
# Cleaning 'Age' column. Under 1 is now 0
data['Age'] = data['Age'].str.replace('under 1', '0')

# Printing
data.head()
```

```python
# Changing 'Age' dtype to int
data['Age'] = data['Age'].astype('int')

# Printing
data.head()
```

```python
# Binning

# Bins
bins=[0,25,50,75,100]
# Labels
bin_labels=["age_0_25","age_26_50", "age_51_75", "age_76_100"]



data["Age_Demographic"] = pd.cut(data["Age"], bins, labels=bin_labels, include_lowest=True)
# Printing DataFrame
data.head()
```
