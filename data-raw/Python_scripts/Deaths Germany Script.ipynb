```python
# Importing libraries
import pandas as pd
import numpy as np
import csv
import matplotlib as plt
from datetime import datetime as dt
import datetime as dt
```

```python
# Reading file
data = pd.read_csv("Deaths_Germany_years_nationality_sex_age_years_copy.csv", encoding= 'unicode_escape')
```

```python
# Printing dataframe
data.head(50)
```

```python
# Renaming columns
data.rename(columns = {"Unnamed: 0" : "Year", "Unnamed: 1" : "Age",
                       "Germans": "Male_Germans", "Foreigners" : "Male Foreigners",
                       "Total" : "Total Male",
                      "Germans.1" : "Female_Germans", "Foreigners.1" : "Female Foreigners",
                       "Total.1" : "Total Fem",
                      "Germans.2" : "Total_Germans", "Foreigners.2" : "Total Foreigners",
                      "Total.2" : "Total"}, inplace = True)

# Printing
data.head(50)
```

```python
# Setting index to 1 
data.index += 1

# Printing
data.head(50)
```

```python
# Selecting columns to analyze
data = data.loc[:, ["Year", "Age", "Male_Germans", "Female_Germans", "Total_Germans"]]

# Printing
data.head(50)
```

```python
# Cleaning 'Age' column
data['Age'] = data['Age'].str.replace('years', '')

# Printing
data.head(50)
```

```python
# Cleaning 'Age' column
data['Age'] = data['Age'].str.replace(' s', '')

# Printing
data.head(50)
```

```python
# Cleaning 'Age' column
data['Age'] = data['Age'].str.replace(' and over', '')

# Printing
data.head(50)
```

```python
len(data)
```

```python
# Deleting all rows with 'Total'
data = data[data.Age != 'Total']

# Printing
data.head(50)
```

```python
# Cleaning 'Age' column
data['Age'] = data['Age'].str.replace(' year', '')

# Printing
data.head(50)
```

```python
# Cleaning 'Age' column. Under 1 is now 0
data['Age'] = data['Age'].str.replace('under 1', '0')

# Printing
data.head(50)
```

```python
# Changing 'Age' dtype to int
data['Age'] = data['Age'].astype('int')

# Printing
data.head(50)
```

```python
# Checking data types
data.dtypes
```

```python
# Filtering rows: 65 and older, 5 and younger
data = data.loc[(data['Age'] >= 65) | (data['Age'] <= 5)]

# Printing
data.head(50)
```

```python
# Changing 'Year' data type to Date type
# '%Y-%m-%d' --> Year- month as number - day of month
data.date = pd.to_datetime(data['Year'], format = '%Y-%m-%d')

# Printing
data.head(50)
```
